```python
# -----------------------------------------
# DO NOT EDIT THIS CELL
# -----------------------------------------
# The only exception is if you need to run "!pip install" to install a package prior to importing
# This cell loads the required Python packages

# Import geospatial libraries
import geodatasets
import geopandas as gpd
from shapely.geometry import Point, LineString, Polygon

# Import visualization and utility libraries
import matplotlib.pyplot as plt
import numpy as np
```

```python
# -----------------------------------------
# DO NOT EDIT THIS CELL
# -----------------------------------------
# This cell shows available geospatial datasets.
# Scroll through the output and locate the one named "natregimes".
# You will use it in the next step.

geodatasets.data
```

```python
# -----------------------------------------
# INSTRUCTIONS: TASK 1
# -----------------------------------------
# Use the "read_file" and "get_path" functions appropriately
# to extract the "natregimes" data into an object named "counties"

# Get the path to the natregimes dataset
natregimes_path = geodatasets.get_path("geoda.natregimes")

# Read the geospatial data into a GeoDataFrame
counties = gpd.read_file(natregimes_path)
```

```python
# -----------------------------------------
# INSTRUCTIONS: TASK 2
# -----------------------------------------
# Print out the first few observations and the columns of "counties"

# Display the column names
print("Columns in the dataset:")
print(counties.columns.tolist())

print("\n\nFirst 5 observations:")
counties.head()
```

```python
# -----------------------------------------
# INSTRUCTIONS: TASK 3
# -----------------------------------------
# Construct a map visualization using the entire dataset in "counties"
# Make the title of the visualization "All US Counties"

# Create a figure and axis for the plot
fig, ax = plt.subplots(1, 1, figsize=(15, 10))

# Plot all US counties
counties.plot(ax=ax, color='lightblue', edgecolor='white', linewidth=0.5)

# Set the title
ax.set_title('All US Counties', fontsize=16, fontweight='bold')

# Remove axis labels for cleaner appearance
ax.set_xlabel('')
ax.set_ylabel('')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# -----------------------------------------
# INSTRUCTIONS: TASK 4
# -----------------------------------------
# Create an object named "massachusetts" that only includes Massachusetts's counties from your "counties" object
# Construct a map visualization of Massachusetts's counties and give the visualization an appropriate title

# Filter for Massachusetts counties
massachusetts = counties[counties['STATE_NAME'] == 'Massachusetts'].copy()

# Create the visualization
fig, ax = plt.subplots(1, 1, figsize=(12, 8))

# Plot Massachusetts counties
massachusetts.plot(ax=ax, color='lightgreen', edgecolor='darkgreen', linewidth=1)

# Set the title
ax.set_title('Massachusetts Counties', fontsize=16, fontweight='bold')

# Remove axis labels for cleaner appearance
ax.set_xlabel('')
ax.set_ylabel('')

# Show the plot
plt.tight_layout()
plt.show()

print(f"\nNumber of Massachusetts counties: {len(massachusetts)}")
```

```python
# -----------------------------------------
# INSTRUCTIONS: TASK 5
# -----------------------------------------
# Create an object named "mississippi" that only includes Mississippi's counties from your "counties" object
# Construct a map visualization of Mississippi's counties and give the visualization an appropriate title

# Filter for Mississippi counties
mississippi = counties[counties['STATE_NAME'] == 'Mississippi'].copy()

# Create the visualization
fig, ax = plt.subplots(1, 1, figsize=(12, 8))

# Plot Mississippi counties
mississippi.plot(ax=ax, color='lightcoral', edgecolor='darkred', linewidth=1)

# Set the title
ax.set_title('Mississippi Counties', fontsize=16, fontweight='bold')

# Remove axis labels for cleaner appearance
ax.set_xlabel('')
ax.set_ylabel('')

# Show the plot
plt.tight_layout()
plt.show()

print(f"Number of Mississippi counties: {len(mississippi)}")
```

```python
# -----------------------------------------
# INSTRUCTIONS: TASK 6
# -----------------------------------------
# Refer to counties columns and choose one numeric measure from 1 of the 5 columns: HR90, UE90, DV90, FP89, MA90
    # HR90: Homicide Rate (1990)
    # UE90: Unemployment Rate (1990)
    # DV90: Divorce Rate (1990)
    # FP89: % of Families Below Poverty Level (1989)
    # MA90: Median Age (1990)
# Use your chosen measure to create two map plots:
    # 1. A map of Massachusetts's counties whose color fills correspond to the value of your chosen measure
    # 2. A map of Mississippi's counties whose color fills correspond to the value of your chosen measure
# HINT: Use the "column" argument in your plot function properly

# I choose UE90 (Unemployment Rate 1990) as the first numeric measure
chosen_measure_1 = 'UE90'

# Create subplots for side-by-side comparison
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 8))

# 1. Massachusetts counties colored by unemployment rate
massachusetts.plot(column=chosen_measure_1, ax=ax1, cmap='OrRd',
                  edgecolor='black', linewidth=0.5, legend=True)
ax1.set_title(f'Massachusetts Counties - {chosen_measure_1} (Unemployment Rate 1990)',
              fontsize=14, fontweight='bold')
ax1.set_xlabel('')
ax1.set_ylabel('')

# 2. Mississippi counties colored by unemployment rate
mississippi.plot(column=chosen_measure_1, ax=ax2, cmap='OrRd',
                edgecolor='black', linewidth=0.5, legend=True)
ax2.set_title(f'Mississippi Counties - {chosen_measure_1} (Unemployment Rate 1990)',
              fontsize=14, fontweight='bold')
ax2.set_xlabel('')
ax2.set_ylabel('')

plt.tight_layout()
plt.show()
```

```python
# -----------------------------------------
# INSTRUCTIONS: TASK 7
# -----------------------------------------
# Recreate your prior two plots from the previous cell, except this time make the color fill scale the same across both maps
# HINT: "vmin" and "vmax" arguments in your plot function can help in this regard

# Calculate the overall min and max values across both states for consistent color scale
overall_min = min(massachusetts[chosen_measure_1].min(), mississippi[chosen_measure_1].min())
overall_max = max(massachusetts[chosen_measure_1].max(), mississippi[chosen_measure_1].max())

print(f"Using consistent color scale: {overall_min:.2f}% to {overall_max:.2f}%")

# Create subplots for side-by-side comparison
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 8))

# 1. Massachusetts counties with consistent color scale
massachusetts.plot(column=chosen_measure_1, ax=ax1, cmap='OrRd',
                  edgecolor='black', linewidth=0.5, legend=True,
                  vmin=overall_min, vmax=overall_max)
ax1.set_title(f'Massachusetts Counties - {chosen_measure_1} (Unemployment Rate 1990)\nConsistent Color Scale',
              fontsize=14, fontweight='bold')
ax1.set_xlabel('')
ax1.set_ylabel('')

# 2. Mississippi counties with consistent color scale
mississippi.plot(column=chosen_measure_1, ax=ax2, cmap='OrRd',
                edgecolor='black', linewidth=0.5, legend=True,
                vmin=overall_min, vmax=overall_max)
ax2.set_title(f'Mississippi Counties - {chosen_measure_1} (Unemployment Rate 1990)\nConsistent Color Scale',
              fontsize=14, fontweight='bold')
ax2.set_xlabel('')
ax2.set_ylabel('')

plt.tight_layout()
plt.show()

print(f"\nWith consistent color scale ({overall_min:.2f}% - {overall_max:.2f}%), ")
print("you can now directly compare unemployment rates between the two states.")
print("Darker colors indicate higher unemployment rates in both maps.")
```

```python
# -----------------------------------------
# INSTRUCTIONS: TASK 8
# -----------------------------------------
# Refer to counties columns and choose a SECOND numeric measure from 1 of the 5 columns: HR90, UE90, DV90, FP89, MA90
    # HR90: Homicide Rate (1990)
    # UE90: Unemployment Rate (1990)
    # DV90: Divorce Rate (1990)
    # FP89: % of Families Below Poverty Level (1989)
    # MA90: Median Age (1990)
# Use your second chosen measure to create two map plots:
    # 1. A map of Massachusetts's counties whose color fills correspond to the value of your chosen measure
    # 2. A map of Mississippi's counties whose color fills correspond to the value of your chosen measure
# Also, the two maps should use the same MINIMUM and MAXIMUM for your color scale.

# I choose FP89 (% of Families Below Poverty Level 1989) as the second numeric measure
chosen_measure_2 = 'FP89'

# Calculate the overall min and max values across both states for consistent color scale
overall_min_2 = min(massachusetts[chosen_measure_2].min(), mississippi[chosen_measure_2].min())
overall_max_2 = max(massachusetts[chosen_measure_2].max(), mississippi[chosen_measure_2].max())

print(f"Second measure: {chosen_measure_2} (% of Families Below Poverty Level 1989)")
print(f"Using consistent color scale: {overall_min_2:.1f}% to {overall_max_2:.1f}%")

# Create subplots for side-by-side comparison
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 8))

# 1. Massachusetts counties colored by poverty rate with consistent scale
massachusetts.plot(column=chosen_measure_2, ax=ax1, cmap='Reds',
                  edgecolor='black', linewidth=0.5, legend=True,
                  vmin=overall_min_2, vmax=overall_max_2)
ax1.set_title(f'Massachusetts Counties - {chosen_measure_2}\n(% Families Below Poverty Level 1989)',
              fontsize=14, fontweight='bold')
ax1.set_xlabel('')
ax1.set_ylabel('')

# 2. Mississippi counties colored by poverty rate with consistent scale
mississippi.plot(column=chosen_measure_2, ax=ax2, cmap='Reds',
                edgecolor='black', linewidth=0.5, legend=True,
                vmin=overall_min_2, vmax=overall_max_2)
ax2.set_title(f'Mississippi Counties - {chosen_measure_2}\n(% Families Below Poverty Level 1989)',
              fontsize=14, fontweight='bold')
ax2.set_xlabel('')
ax2.set_ylabel('')

plt.tight_layout()
plt.show()

print(f"\nWith consistent color scale ({overall_min_2:.1f}% - {overall_max_2:.1f}%), ")
print("you can directly compare poverty levels between Massachusetts and Mississippi.")
print("Darker red colors indicate higher poverty rates in both maps.")
```
