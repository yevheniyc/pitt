```python
# Part 1: Matplot Fundamentals
# Essential imports for data visualization
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set style for better visualizations
plt.style.use('seaborn-v0_8')  # Updated to use correct style name

# Turn on inline plotting for Jupyter
%matplotlib inline
```

```python
# 1. Pyplot Style (Simple but less flexible)
plt.figure(figsize=(10, 5))
x = np.linspace(0, 10, 100)
plt.plot(x, np.sin(x))
plt.title('Sine Wave using Pyplot Style')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.grid(True)
plt.show()
```

```python
# 2. Object-Oriented Style (More flexible, recommended)
fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(x, np.sin(x))
ax.set_title('Sine Wave using Object-Oriented Style')
ax.set_xlabel('x')
ax.set_ylabel('sin(x)')
ax.grid(True)
plt.show()
```

```python
# Create multiple subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))

# First subplot: Sine wave
x = np.linspace(0, 10, 100)
ax1.plot(x, np.sin(x), label='sin(x)')
ax1.set_title('Sine Wave')
ax1.set_xlabel('x')
ax1.set_ylabel('sin(x)')
ax1.grid(True)
ax1.legend()

# Second subplot: Cosine wave
ax2.plot(x, np.cos(x), color='red', label='cos(x)')
ax2.set_title('Cosine Wave')
ax2.set_xlabel('x')
ax2.set_ylabel('cos(x)')
ax2.grid(True)
ax2.legend()

# Adjust layout to prevent overlap
plt.tight_layout()
plt.show()
```

```python
# Understanding Figure and Aces Objects

# Create a more complex layout
fig = plt.figure(figsize=(12, 8)) # Create a figure with 12 inches wide and 8 inches tall size

# Add Suplotis in specific positions
ax1 = fig.add_subplot(2, 2, 1) # 2 rows, 2 columns, first subplot
ax2 = fig.add_subplot(2, 2, 2) # 2 rows, 2 columns, second subplot
ax3 = fig.add_subplot(2, 2, (3, 4)) # 2 rows, 2 columns, third and fourth subplots

# Plot different types of data

# Suplot 1: Line plot
x = np.linspace(0, 5, 50)
ax1.plot(x, x**2, 'b-', label='x^2') # 'b-' specifies a blue solid line
ax1.set_title('Square Function')
ax1.legend()

# Suplot 2: Scatter plot
np.random.seed(2100)
x = np.random.rand(50)
y = np.random.rand(50)
ax2.scatter(x, y, c='red', alpha=0.5)
ax2.set_title('Random Scatter Plot')

# Suplot 3: Horizontal Bar Chart
categories = ['A', 'B', 'C', 'D']
values = [3, 7, 5, 4]
ax3.barh(categories, values, color='green')
ax3.set_title('Horizontal Bar Chart')

plt.tight_layout()
plt.show()


x
```

```python
# Part 2: Single Variable Visualization

"""
1. Data type determines plot type:
- Categorical: Bar, Pie, Countplot
- Continuous: Histogram, Boxplot, Scatterplot

2. Number of unique values matters
3. We will focus on marginal plots (single variables)

"""
```

```python
# Load the penguins dataset
penguins = sns.load_dataset('penguins')

print('Dataset Overview')
print(penguins.info())
penguins.head()
```

```python
# Bar Charts for categorical variables
fig, axes = plt.subplots(2, 2, figsize=(15, 10)) # Create a 2x2 grid of subplots
axes = axes.ravel() # Flatten the axes array to make it easier to iterate over

# 1. Species Distribution
species_counts = penguins['species'].value_counts()
axes[0].bar(species_counts.index, species_counts.values)
axes[0].set_title('Penguin Species Distribution')
axes[0].set_xlabel('Species')
axes[0].set_ylabel('Count')

# 2. Sex Distribution
sex_counts = penguins['sex'].value_counts()
axes[1].bar(sex_counts.index, sex_counts.values, color='salmon')
axes[1].set_title('Penguin Sex Distribution')
axes[1].set_xlabel('Sex')
axes[1].set_ylabel('Count')

# 3. Island Distribution
island_counts = penguins['island'].value_counts()
axes[2].bar(island_counts.index, island_counts.values, color='lightgreen')
axes[2].set_title('Penguin Island Distribution')
axes[2].set_xlabel('Island')
axes[2].set_ylabel('Count')

# 4 Same Data but as proportions
island_props = penguins['island'].value_counts(normalize=True)
axes[3].bar(island_props.index, island_props.values * 100, color='lightblue')
axes[3].set_title('Penguin Island Distribution (%)')
axes[3].set_xlabel('Island')
axes[3].set_ylabel('Percentage')

plt.tight_layout()
plt.show()
```

```python
# Continuous Variables Visualization: Histograms

fig, axes = plt.subplots(2, 2, figsize=(15, 10))
axes = axes.ravel()

# 1. Basic histogram
axes[0].hist(penguins['bill_length_mm'].dropna(), bins=20) # bins = number of bars
axes[0].set_title('Histogram of Bill Length')
axes[0].set_xlabel('Bill Length (mm)')
axes[0].set_ylabel('Frequency')

# 2. History with KDE - Kernel Density Estimation which is a smoother version of the histogram
sns.histplot(data=penguins, x="flipper_length_mm", kde=True, ax=axes[1])
axes[1].set_title('Flipper Length Histogram with KDE')

# 3. KDE plot
sns.kdeplot(data=penguins, x="body_mass_g", ax=axes[2], fill=True) # fill=True makes the area under the curve
axes[2].set_title('KDE Plot of Body Mass')
axes[2].set_xlabel('Body Mass (g)')
axes[2].set_ylabel('Density')


# 4. Plot multiple KDEs for different species
for species in penguins['species'].unique():
    subset = penguins[penguins['species'] == species]
    sns.kdeplot(data=subset, x='bill_depth_mm', ax=axes[3], label=species)
axes[3].set_title('KDE Plot of Bill Depth by Species')
axes[3].set_xlabel('Bill Depth (mm)')
axes[3].set_ylabel('Density')
axes[3].legend()

plt.tight_layout()
plt.show()
```

```python
# Example of treating a continuous variable as categorical

# Create a new variable: rounded body mass in kg
penguins['body_mass_kg'] = (penguins['body_mass_g'] / 1000).round()

# Create a bar chart of this discrete numeric variable - discrete means it's not continuous
plt.figure(figsize=(10, 6))
mass_counts = penguins['body_mass_kg'].value_counts().sort_index()
plt.bar(mass_counts.index, mass_counts.values)
plt.title('Body Mass Distribution')
plt.xlabel('Body Mass (kg)')
plt.ylabel('Frequency')
print(mass_counts.head())
plt.show()
```

```python
# Compare different visualization approaches
fig, axes = plt.subplots(2, 2, figsize=(15, 10))
axes = axes.ravel()

# 1. Histogram with few bins
sns.histplot(data=penguins, x='bill_length_mm', bins=5, ax=axes[0])
axes[0].set_title('Histogram with 5 bins')

# 2. Histogram with many bins
sns.histplot(data=penguins, x='bill_length_mm', bins=50, ax=axes[1])
axes[1].set_title('Histogram with 50 bins')

# 3. KDE only
sns.kdeplot(data=penguins, x='bill_length_mm', ax=axes[2])
axes[2].set_title('KDE only')

# 4. Combined histogram and KDE
sns.histplot(data=penguins, x='bill_length_mm', kde=True, bins=21, ax=axes[3])
axes[3].set_title('Histogram with KDE (21 bins)')

plt.tight_layout()
plt.show()
```

```python
# Part 3: Two-Variable Relationships:

"""
1. Categorical vs Caterical Relationships
2. Categorical vs Continuous Relationships
"""

penguins = sns.load_dataset('penguins')
```

```python
# 1. Using cross-tabulation to explore categorical vs categorical relationships
species_island_cross = pd.crosstab(penguins['species'], penguins['island'])
print("Cross-tabulation of Species vs Island:")
print(species_island_cross)

# Visualizing these relationships in different ways

# 1. Stacked Bar Chart
plt.figure(figsize=(12, 5))
species_island_cross.plot(kind='bar', stacked=True)
plt.title('Species Distribution by Island')
plt.xlabel('Species')
plt.ylabel('Count')
plt.legend(title='Island')
plt.tight_layout()
plt.show()
```

```python
# 2. Grouped Bar Chart Using Seaborn
plt.figure(figsize=(12, 5))
sns.catplot(data=penguins, x='species', hue='island', kind='count', height=6, aspect=2)
# height = height of the plot
# aspect = aspect ratio of the plot - which is width/height, where 2 is the width and 1 is the height
# kind='count' means that we are using a count plot, others are bar, point, line, etc.
# hue = color the bars by the island variable
# x = the variable to display on the x-axis
# y = the variable to display on the y-axis, in this case, it's the count

plt.title('Species Distribution by Island')
plt.xlabel('Species')
plt.ylabel('Count')
plt.tight_layout()
plt.show()

```

```python
# 3. Heatmap - useful for visualizing the relationship between two categorical variables
plt.figure(figsize=(10, 6))
sns.heatmap(species_island_cross, annot=True, fmt='d', cmap='YlOrRd')
# annot = True means that we are annotating the cells with the values
# fmt = 'd' means that we are formatting the values as integers
# cmap = 'YlOrRd' means that we are using the YlOrRd colormap, which is a yellow to red colormap
plt.title('Species Distribution by Island')
plt.xlabel('Island')
plt.ylabel('Species')
plt.tight_layout()
plt.show()

# 4. Pairplot - useful for visualizing the relationship between two continuous variables
sns.pairplot(penguins)
plt.show()
```

```python
# Categorical vs Continuous Relationships

# Create a figure with multiple visualization types
fig, axes = plt.subplots(2, 2, figsize=(15, 12))
axes = axes.ravel()

# 1. Boxplot
sns.boxplot(data=penguins, x='species', y='bill_length_mm', ax=axes[0])
axes[0].set_title('Boxplot of Bill Length by Species')
axes[0].set_xlabel('Species')
axes[0].set_ylabel('Bill Length (mm)')

# 2. Violin Plot
sns.violinplot(data=penguins, x='species', y='bill_length_mm', ax=axes[1])
axes[1].set_title('Violin Plot of Bill Length by Species')
axes[1].set_xlabel('Species')
axes[1].set_ylabel('Bill Length (mm)')

# 3. Strip Plot (Individual observations)
sns.stripplot(data=penguins, x='species', y='bill_length_mm', ax=axes[2])
axes[2].set_title('Strip Plot of Bill Length by Species')
axes[2].set_xlabel('Species')
axes[2].set_ylabel('Bill Length (mm)')

# 4. Combined Box and Strip Plots
sns.boxplot(data=penguins, x='species', y='bill_length_mm', ax=axes[3])
sns.stripplot(data=penguins, x='species', y='bill_length_mm', ax=axes[3], color='red', alpha=0.5)
axes[3].set_title('Combined Box and Strip Plots of Bill Length by Species')
axes[3].set_xlabel('Species')
axes[3].set_ylabel('Bill Length (mm)')

plt.tight_layout()
plt.show()
```

```python
# Faceted histograms - useful for visualizing the relationship between a continuous variable and a categorical variable

g = sns.FacetGrid(penguins, col='species', height=5, aspect=.8)
# aspect = width/height, which means the width is 0.8 times the height
g.map(sns.histplot, 'bill_length_mm')
plt.show()
```

```python
# Let's analyze Gapminder dataset as per our previous discussion and what has been covered in the lectures
# Load Gapminder dataset from URL
url="https://raw.githubusercontent.com/CODATA-RDA-DataScienceSchools/Materials/master/docs/DataSanJose2019/slides/Visualisation/gapminder.csv"
gapminder=pd.read_csv(url)

# Basic info about the dataset
print(gapminder.info())
print(gapminder.head())
```

```python
# Initial Distribution of Life Expectancy
# This chart shows the distribution of life expectancy in the dataset
# which represents how long people live and how many people live in that age range
# why group in 70 year range shows the highest frequency?
# Because there are more people who live in that age range and the reason is because of the population growth
plt.figure(figsize=(10, 6))
sns.histplot(gapminder['lifeExp'], bins=30, kde=True)
plt.title('Distribution of Life Expectancy')
plt.xlabel('Life Expectancy')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
# Conditional KDE per year
plt.figure(figsize=(12, 8))
sns.kdeplot(data=gapminder, x='lifeExp', hue='year', fill=True)
plt.title('Conditional KDE of Life Expectancy by Year')
plt.xlabel('Life Expectancy')
plt.ylabel('Density')
plt.tight_layout()
```

```python
# 3. Box plot for summary statistics
plt.figure(figsize=(12, 8))
sns.boxplot(data=gapminder, x='year', y='lifeExp')
plt.title('Box Plot of Life Expectancy by Year')
plt.xlabel('Year')
plt.ylabel('Life Expectancy')
plt.tight_layout()
```

```python
# 4. Point plot for average changes -> shows confidence intervals
plt.figure(figsize=(12, 8))
sns.pointplot(data=gapminder, x='year', y='lifeExp')
plt.title('Point Plot of Life Expectancy by Year')
plt.xlabel('Year')
plt.ylabel('Life Expectancy')
plt.tight_layout()
```

```python
# 5. Final analysis with continent
sns.catplot(data=gapminder,
            x='year',
            y='lifeExp',
            hue='continent',
            kind='box',
            aspect=2)
plt.show()
```
